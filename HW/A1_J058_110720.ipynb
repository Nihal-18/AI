```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
class Reading_Files(object):
  def __init__(self, fname):
    self.fname = fname

  def read_rows(self):
    df = pd.read_csv(self.fname)
    for i in range(2):
      r = np.random.randint(0,100)
      x = df.iloc[r,:]
      print('\n\nRow {} contents are \n{}'.format(r,x))
```

```
class DataRead2(Reading_Files):
  def __init__(self,fname):
    self.fname = fname

  def corr(self):
    df = pd.read_csv(self.fname)

    import seaborn as sb

    corr = df.corr()

    sb.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
    plt.figure(figsize=(13,7))
    a = sb.heatmap(corr, annot=True, fmt='.2f')
    rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
    roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)

  def override(self):
    df1 = pd.read_csv(self.fname)
    for i in range(3):
      r = np.random.randint(0,100)
      x = df1.iloc[r,:]
      print('\n\nRow {} contents are \n{}'.format(r,x))

```

```
d1 = Reading_Files('heart.csv')
```
